```python
from collections import defaultdict ###Import stuff and make dictionaries

f = open("C:/Users/jbro0/OneDrive/Desktop/College_Stuff/PoCSVerse/Project/Github Stuff/Pocs_Film_Project/Data/movies-vs-tropes.csv")
tropes_dict = defaultdict(list)
for line in f:
    trope_string = line.split(";")[1] ### Find tropes associated with movie
    indiv_tropes = trope_string.split(",") ### Seperate into individual tropes
    movie = line.split(";")[0]
    for trope in indiv_tropes:
        tropes_dict[trope].append(movie)
f.close()

for trope in list(tropes_dict.keys()): ### Make values lists 
    movies = tropes_dict[trope]
    tropes_dict[trope] = movies
```

```python
test_trope = "ThoseTwoGuys"
with open("C:/Users/jbro0/OneDrive/Desktop/College_Stuff/PoCSVerse/Project/Github Stuff/Pocs_Film_Project/Data/liteweight/genre_counts_by_trope.csv", "r") as file:
    for line in file:
        if line.split(",")[0] == "":
            genres = list(line.split(","))
            genres.pop(0)
            genres.pop(0)
            genres.pop(0)
            
        if line.split(",")[1] == test_trope:
            amounts = list(line.split(","))
            amounts.pop(0)
            amounts.pop(0)
            amounts.pop(0)
            

genres[-1] = genres[-1].replace("\n", "")
amounts[-1] = amounts[-1].replace("\n", "")
genre_counts = [int(amount) for amount in amounts]
print(genre_counts)
print(genres)

genre_percents = [genre_count / sum(genre_counts) for genre_count in genre_counts]

        
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def color_list():
    random_colors = []
    for j in range(0, 19):
        # Generate random RGB values
        r = np.random.rand()
        g = np.random.rand()
        b = np.random.rand()

        # Create the color tuple
        random_colors.append([r, g, b])
    return(random_colors)


fig, ax = plt.subplots()

ax.bar(x = [i for i in range(0,19)], height=genre_percents, label = genres, color = color_list())
ax.legend(title = "Genres", fontsize="x-small")
ax.set_ylabel("Counts")
####ax.set_yticks([i for i in range(0, max(genre_counts)+1)])
ax.set_xlabel("Arbitrary")
ax.set_title("Histogram of genre percentages for the " + str(test_trope) + " trope")


```
