```python
import pandas as pd
import os 
from Utility.toolbox import find_repo_root

root = find_repo_root()
```

```python
df = pd.read_csv(f"{root}/Data/Whole_sets/2020_updated_tropeset.csv")
df.head()
```

```python
df.columns
```

```python
genre_columns = df.filter(like='GENRE', axis=1).columns
genre_columns = genre_columns.tolist() 
movie_columns =[ 'Id', 'NameTvTropes', 'NameIMDB', 'Rating', 'Votes', 'Year']
exclude_columns = genre_columns + movie_columns
tropes = df.drop(columns=exclude_columns).columns
tropes.tolist()
print("")
```

```python
result = []

for trope in tropes:
    # sum up movies  w trope (value = 1)
    genre_sums = df[df[trope] == 1][genre_columns].sum()
    
    # Collect associated movies
    associated_movies = df[df[trope] == 1]['NameIMDB'].tolist()

    # put into tuple. * expands the sum object into lots of entries in tuple 
    ret = (trope, *genre_sums, associated_movies)

    result.append(ret)


result_df = pd.DataFrame(result, columns=['Trope'] + genre_columns +['Associated_Movies'])
result_df
```

```python
result_df['Number_movies'] = result_df['Associated_Movies'].apply(len)
result_df.columns = [col.replace("GENRE", "") if "GENRE" in col else col for col in result_df.columns]
result_df.columns = [col.replace("[]", "") if "[]" in col else col for col in result_df.columns]
result_df
```

```python
result_df.to_csv(f"{root}/Data/liteweight/2020_genre_counts_by_trope.csv")
```
