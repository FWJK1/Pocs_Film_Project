```python
import os

def find_repo_root(start_path):
    """
    useful general function for finding the (first, closest) repo root so github file paths work the same on different machines 
    """
    current_path = os.path.abspath(start_path)
    
    while True:
        # Check for the existence of the .git directory or other indicators
        if os.path.isdir(os.path.join(current_path, '.git')) or \
           os.path.isfile(os.path.join(current_path, 'README.md')):
            return current_path
        
        parent_path = os.path.dirname(current_path)
        
        # Stop if we reach the root directory
        if parent_path == current_path:
            break
        
        current_path = parent_path

    return None  # Return None if not found

root = find_repo_root(os.getcwd())
print(root)
```

```python
def strip_year(text):
    if pd.isna(text):
        print(text)
        return text # Return NaN and None if text is NaN
    
    match = pd.Series(text).str.extract(r'(\s(\d{4}))$')
    if not match.empty:
        return text.replace(str(match[0][0]), '').strip()
    return text


def get_year(text):
    if pd.isna(text):
        return None # Return NaN and None if text is NaN
    
    match = pd.Series(text).str.extract(r'(\s(\d{4}))$')
    if not match.empty:
        return match[1][0]
    return None

df_tropes['Movie_strip'] = df_tropes['Movie'].map(lambda x: strip_year(x))
df_tropes['year'] = df_tropes['Movie'].map(lambda x: get_year(x))

```

```python
imdb_link =imdb.IMDb()

def search_movie(link, title, year=None):
    if pd.isna(year):
        year = None
    else:
        year = str(year)
        # print(year)
    
    # Search for movies with the given title
    try:
        results = link.search_movie(title)
    except Exception as e:
        print(e)
        return None

    if not results:  # Check if there are any results
        print("No results found for:", title)
        return None
    
    # Filter results by year
    else: 
        if year:
            for movie in results:
                # print(movie)
                if movie.get('year') == year:
                    return movie.movieID # Return the movie data if found
        return results[0].movieID
```

```python
imdb_link =imdb.IMDb()
df_tropes.columns  = df_tropes.columns.str.strip()
## see most recent printout to know where to start 
chunk_size = 100
for start in range(10800, 10901, chunk_size):
    end = start + chunk_size
    print(f"starting {start} to {end}")
    chunk = df_tropes.iloc[start:end].copy()

    # Apply the search_movie function to the chunk
    chunk['imdb_'] = chunk.apply(
        lambda row: search_movie(imdb_link, row['Movie_strip'], row['year']),
        axis=1
    )
    
    # Save the chunk to a CSV or append to an existing one
    chunk.to_csv(f'{root}/Data/tropes_with_imfb.csv', mode='a', header=not pd.io.common.file_exists('results.csv'), index=False)
    print(f"completed {start} to {end} \n\n")
```
