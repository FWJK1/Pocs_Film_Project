```python
import pandas as pd
import re
import os 
import ast
import spacy
```

```python
def find_repo_root(start_path):
    """
    useful general function for finding the (first, closest) repo root so github file paths work the same on different machines 
    """
    current_path = os.path.abspath(start_path)
    
    while True:
        # Check for the existence of the .git directory or other indicators
        if os.path.isdir(os.path.join(current_path, '.git')) or \
           os.path.isfile(os.path.join(current_path, 'README.md')):
            return current_path
        
        parent_path = os.path.dirname(current_path)
        
        # Stop if we reach the root directory
        if parent_path == current_path:
            break
        
        current_path = parent_path

    return None  # Return None if not found

root = find_repo_root(os.getcwd())

root = root.replace('\\', '/')
```

```python
search_movie = pd.read_csv(f"{root}/Data/tropes_year_movie.csv")
search_movie
```

```python
tropes_year_movie = pd.read_csv(f"{root}/Data/letterboxd_search.csv")
tropes_year_movie.drop(columns=['Unnamed: 0'])
```

```python
df = pd.merge(
    search_movie,
    tropes_year_movie,
    on="letterboxd_search"
)
df = df[['letterboxd_search', 'Tropes']]
df.to_csv(f"{root}/Data/Whole_sets/search_tropes_for_merging.csv")
```
